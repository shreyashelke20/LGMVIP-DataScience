# Beginner Level Task 2: Image to Pencil Sketch with Python 

# 1.Importing the necessary libraries:
              We have imported all the required libraries as well as mounted the drive to upload the image that we will be converting to sketch.

```python
import numpy as np
import pandas as pd
import scipy.ndimage
import cv2 
from google.colab.patches import cv2_imshow
from skimage import io
from PIL import Image
import matplotlib.pylab as plt
from google.colab import drive
drive.mount('/content/drive')
```

Next, we will read the image.

```python
img = io.imread('/content/drive/MyDrive/cat.jpg')
cv2_imshow(img)
```

As we can see, the image is in BGR format. Let us convert into RGB to view the original image.

```python
new= cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
Image.fromarray(new).save('/content/drive/MyDrive/cat.jpg')
cv2_imshow(new)
```

# 2. Convert Image to a Pencil Sketch by performing the following steps:

Convert to gray image:
                     

```python
gray = cv2.cvtColor(new, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

Invert image:

```python
negative=cv2.bitwise_not(gray)
cv2_imshow(negative)
```

Blur Image:

```python
blurred=cv2.GaussianBlur(negative, (25,25),0)
cv2_imshow(blurred)
```

Invert blurred image:

```python
inv=cv2.bitwise_not(blurred)
cv2_imshow(inv)
```

Sketch:

```python
sketch=cv2.divide(gray,inv, scale=256.0)
cv2.imwrite('sketch.png', sketch)
cv2_imshow(sketch)
```

Thus, we have successfully converted an image into a sketch. Let us compare the two side by side.

```python
plt.figure(figsize=(14,8))
plt.subplot(1,2,1)
plt.title('Original image', size=18)
original= cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
Image.fromarray(new).save('/content/drive/MyDrive/cat.jpg')
plt.imshow(original)
plt.axis('off')
plt.subplot(1,2,2)
plt.title('Sketch', size=18)
rgb_sketch=cv2.cvtColor(sketch, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_sketch)
plt.axis('off')
plt.show()
```
